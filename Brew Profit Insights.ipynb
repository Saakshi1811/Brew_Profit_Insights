```python
def proper_df(df):

  df.columns=[i.lower().replace(' (₹)','').replace(' ','_').replace('.','').replace('@','')\
              .replace('!','').replace('#','').replace('$','').replace('%','').replace('&','')\
              .replace('(','').replace(')','') for i in df.columns]

  df.drop([i for i in df.columns if df[i].isna().sum()==len(df)],axis=1,inplace=True)

  newdf = pd.DataFrame()

  dict1={i:df[i].isna().sum() for i in df.columns if df[i].isna().sum()!=0}

  newdf['colname'] = dict1.keys()
  newdf['nullcount'] = dict1.values()
  newdf['valuecounts'] = [df[i].value_counts().to_dict() for i in dict1.keys()]

  return newdf

  
```
